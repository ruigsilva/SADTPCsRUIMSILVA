```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Oct 22 00:30:08 2018

@author: Rui Guilherme Silva 19901535
            Informática de Gestão
            TPC01 SAD
"""

def solution(expression):
    if len(expression) % 2 != 0:
        return 0

    opening = ("(", "[", "{")  #estabelece os carateres de abertura
    closing = (")", "]", "}")   #estabelece os carateres de fecho
    mapping = {opening[0]:closing[0],  #mapeia a relação de abertura com o fecho
               opening[1]:closing[1],
               opening[2]:closing[2]}

    if expression[0] in closing:
        return 0

    if expression[-1] in opening:
        return 0

    closing_queue = []
    for letter in expression:
        if letter in opening:
            closing_queue.append(mapping[letter])
        elif letter in closing:
            if not closing_queue:
                return 0

            if closing_queue[-1] == letter:
                closing_queue.pop()
            else:
                return 0
    return 1

# a rotina seguinte lança no ecra a pergunta para introdução da string
print ()
print ('Por favor introduza uma string ?')
intro = input()
print ('resultado da verifcação da string é: ')
print (solution (intro))

#alternativa sem questão pode ser usar o seguinte:
#print ()
#print (solution('(((({{}}))))))'))


```
